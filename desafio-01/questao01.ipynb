#### Questão 01 - Calcule os seguintes itens
a) Número total de palavras 
b) Número total de sentenças
c) Número total de palavras não repetidas
d) Número total de palavras repetidas 
e) Média de palavras por sentenças

1. Criei uma função que faz o cálculo de todos os itens, passando por parâmetro apenas o nome do texto

```python
def calcular_estatisticas(texto):
    # Abre o texto e faz a sua leitura (r)
    with open(texto, 'r', encoding='utf-8', errors='ignore') as file:
        text = file.read()

    # Fiz a "tokenização" das palavras e sentenças
    palavras = nltk.word_tokenize(text)
    sentencas = nltk.sent_tokenize(text)

    # Calcular estatísticas
    num_palavras_total = len(palavras)
    num_sentencas_total = len(sentencas)
    num_palavras_nao_repetidas = len(set(palavras))
    num_palavras_repetidas = num_palavras_total - num_palavras_nao_repetidas
    media_palavras_por_sentenca = num_palavras_total / num_sentencas_total

    # Imprimir resultados
    print(f"Para o texto {texto}:")
    print(f"a) Número total de palavras: {num_palavras_total}")
    print(f"b) Número total de sentenças: {num_sentencas_total}")
    print(f"c) Número total de palavras não repetidas: {num_palavras_nao_repetidas}")
    print(f"d) Número total de palavras repetidas: {num_palavras_repetidas}")
    print(f"e) Média de palavras por sentenças: {media_palavras_por_sentenca:.2f}\n")
```

2. Percorri o vetor de textos para realizar a consulta em cada texto

```python
for texto in textos:
    calcular_estatisticas(nltk.data.find(f'corpora/gutenberg/{texto}'))
```

#### Questão 02 - Em relação ao corpus “gutenberg” implemente os algoritmos para responder às seguintes questões:

a) Total de palavras em cada documento do corpus "gutenberg"

```python
def palavrasPorDocumento():
    for documento in gutenberg.fileids():
        palavras = gutenberg.words(documento)
        total_palavras = len(palavras)
        print(f"Total de palavras em {documento}: {total_palavras}")

palavrasPorDocumento()
```

b) Quem é o maior documento do corpus?


```python
def maiorDocumentoDoCorpus():
    maior_texto = float('-inf') # defini uma váriavel que vai receber menos infinito
    for documento in gutenberg.fileids():
        palavras = gutenberg.words(documento)
        total_palavras = len(palavras)
        if total_palavras > maior_texto:
            maior_texto = total_palavras
            nome_documento = documento
    print(f"O maior documento é {nome_documento} com {maior_texto} palavras.")

maiorDocumentoDoCorpus()
```

c) Quem é o menor documento do corpus?

```python
def menorDocumentoDoCorpus():
    maior_texto = float('+inf') # defini uma váriavel que vai receber mais infinito
    for documento in gutenberg.fileids():
        palavras = gutenberg.words(documento)
        total_palavras = len(palavras)
        if total_palavras < maior_texto:
            maior_texto = total_palavras
            nome_documento = documento
    print(f"O menor documento é {nome_documento} com {maior_texto} palavras.")

menorDocumentoDoCorpus()
```

d) Calcular a média da quantidade de sentenças por palavras do corpus "gutenberg"

```python
def media_sentencas_por_palavras():
    total_sentencas = sum(len(gutenberg.sents(documento)) for documento in gutenberg.fileids())
    total_palavras = sum(len(gutenberg.words(documento)) for documento in gutenberg.fileids())
    media = total_sentencas / total_palavras
    print(f"Média de sentenças por palavra no corpus gutenberg: {media:.2f}")

media_sentencas_por_palavras()
```

e) Calcule a distribuição de frequência das palavras do livro "shakespeare-macbeth.txt"

```python
macbeth_palavras = gutenberg.words('shakespeare-macbeth.txt')
freq_dist = FreqDist(macbeth_palavras)
print(f"Distribuição de frequência das palavras em Macbeth: {freq_dist}")
```

f) Calcule as 5 palavras mais frequentes no corpus "gutenberg"

```python
def cinco_palavras_mais_frequentes():
    todas_palavras = gutenberg.words()
    freq_dist = FreqDist(todas_palavras)
    mais_frequentes = freq_dist.most_common(5)
    print("As 5 palavras mais frequentes no corpus gutenberg são:")
    for palavra, frequencia in mais_frequentes:
        print(f"{palavra}: {frequencia} vezes")

cinco_palavras_mais_frequentes()
```

g) Mostre a diferença de palavras entre dois livros ("shakespeare-caesar.txt" e "shakespeare-hamlet.txt")

```python
def diferenca_entre_dois_livros(livro1, livro2):
    palavras_livro1 = set(gutenberg.words(livro1))
    palavras_livro2 = set(gutenberg.words(livro2))
    diferenca = palavras_livro1.symmetric_difference(palavras_livro2)
    print(f"Quantidade de palavras diferentes entre {livro1} e {livro2}: {len(diferenca)} palavras.")
    print(f"Palavras diferentes: {diferenca}")

diferenca_entre_dois_livros('shakespeare-caesar.txt', 'shakespeare-hamlet.txt')
```

#### Questão 03 - Para o corpus “shakespeare-caesar.txt”, usando expressões regulares responda os seguintes itens:

```python
qtd_err_ocorrencias = len(re.findall(r'err', caesar_texto, flags=re.IGNORECASE))
print(f"Quantidade: {qtd_err_ocorrencias}")
```

#### Questão 04 - Em relação do corpus “shakespeare-hamlet.txt”, faça as seguintes atividades:

```python
hamlet_texto = ' '.join([palavra.lower() for palavra in word_tokenize(hamlet_texto) if palavra.isalpha()])
print(hamlet_texto)
```

c) Aplique o tokenizador em todas as sentenças do corpus;

```python
sentencas = sent_tokenize(hamlet_texto)
print(sentencas)
```

d) Aplique os pos tagger e responda a quantidade de adjetivos existem no corpus; Dica a tag de adjetivo é “JJ”

```python
pos_tags = pos_tag(word_tokenize(hamlet_texto))
qtd_adjetivos = len([tag for palavra, tag in pos_tags if tag == 'JJ'])
print(f"Quantidade: {qtd_adjetivos} adjetivos")
```

#### Questão 05 - Retire as stopwords dos seguintes textos:

1. Carregando os textos que vão ser utilizados:

```python
texto_caesar = gutenberg.raw('shakespeare-caesar.txt')
texto_hamlet = gutenberg.raw('shakespeare-hamlet.txt')
```

2. Função para remover as stop_words e contar a quantidade restante

```python
def remover_stopwords(texto):
    palavras = word_tokenize(texto)
    palavras_sem_stopwords = [palavra for palavra in palavras if palavra.lower() not in stopwords.words('english')]
    qtd_palavras_restantes = len(palavras_sem_stopwords)

    return palavras_sem_stopwords, qtd_palavras_restantes
```

3. Exibindo os dados dos textos

```python
def processar_texto(texto, nome_texto):
    palavras_sem_stopwords, qtd_palavras_restantes = remover_stopwords(texto)

    print(f"\nTexto: {nome_texto}")
    print(f"Palavras Restantes: {qtd_palavras_restantes}")
    print("Exemplo de algumas palavras restantes:")
    print(palavras_sem_stopwords[:10]) 
```

4. Chamando a função de processamento do texto e exibindo os resultados

```python
processar_texto(texto_caesar, 'shakespeare-caesar.txt')
processar_texto(texto_hamlet, 'shakespeare-hamlet.txt')
```
